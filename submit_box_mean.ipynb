```python
EXP   = 'piControl'
var   = 'tos'
#var   = 'thetao'
vtype = 'Omon'
```

```python
dataBen = '/gws/nopw/j04/acsis/bjharvey/storylines/data/JetDiags/'
BenType      = '0to60W'

datadir = '/gws/nopw/j04/acsis/jmecking/'
rundir  = '/work/scratch-nopw/jmecking001/'
```

```python
snames = ['AMV', 'W_S', 'S_S', 'W_M', 'S_M', 'W_N', 'S_N', 'W_S_w', 'S_S_w', 'W_M_w', 'S_M_w', 'W_N_w', 'S_N_w']
imins  = [-75,   -40,   -60,   -80,   -60,   -60,   -60,   -50,     -60,     -80,      -60,    -60,      -45]
imaxs  = [0,     -10,   -10,   -30,   -20,   -20,   -20,   -10,     -20,     -30,      -25,    -20,      0]
jmins  = [0,     10,    20,    25,    35,    45,    50,    10,      15,      25,       30,     40,       45]
jmaxs  = [60,    35,    35,    50,    50,    70,    70,    25,      30,      35,       45,     60,       60]

ns = len(snames)
```

```python
# Loop through all regions:
for ss in range(0,ns):
    sname = snames[ss]
    imin  = imins[ss]
    imax  = imaxs[ss]
    jmin  = jmins[ss]
    jmax  = jmaxs[ss]
    print(sname)

    # Loop through all files and compute if it hasn't be computed yet:
    Files = glob.glob(dataBen + '*' + EXP + '*' + BenType + '_fromseasmean.nc')
    for file in Files:
        Model = Clist[file.split('_')[2]]
        ENS   = file.split('_')[4]
        print(Model.name)
        if ((Model.Ogrid != 'Unstructured') & (Model.Ogrid != 'Unclear') & (Model.Ogrid != 'unknown') & (Model.Ogrid != 'None')):
            modeldir = (datadir + 'CMIP' + Model.cmip + '/' + Model.name + '/' + EXP + '/')
            outfile  = (modeldir + var + '_' +  str(imin) + 'to' + str(imax) + 'E_' + str(jmin) + 'to' + str(jmax) + 'N_Omon_' + Model.name + '_' + EXP + '_' + ENS + '.nc')
            runfile  = (rundir + 'CMIP' + Model.cmip + '/' + var + '_' +  str(imin) + 'to' + str(imax) + 'E_' + str(jmin) + 'to' + str(jmax) + 'N_Omon_' + Model.name + '_' + EXP + '_' + ENS + '.nc')
            if not os.path.isfile(outfile):
                varFiles = Model.getFiles(var,EXP=EXP,ENS=ENS,vtype=vtype,gtype='gn')
                if (len(varFiles) != 0):
                    # Generate submit script for lotus:
                    lfile = ('slurm_files/' + sname + '_' + var + '_' + Model.name + '_' +  EXP + '_' + ENS + '.slurm')
                    line1 = '#!/bin/bash \n'
                    line2 = '#SBATCH --account=short4hr \n'
                    line3 = '#SBATCH -p short-serial-4hr \n'
                    line4 = '#SBATCH -o %J.out \n'
                    line5 = '#SBATCH -e %J.err \n'
                    line6 = '#SBATCH --time 2:00:00 \n'
                    line7 = '\n'
                    line8 = 'conda activate working \n'
                    line9 = ('python ../compute_box_mean_nomask.py ' + Model.name + ' ' + EXP + ' ' + ENS + ' ' + var + ' ' + vtype + ' ' + 
                            str(imin) + ' ' + str(imax) + ' ' + str(jmin) + ' ' + str(jmax) + ' ' + outfile + ' ' + lfile + ' ' + runfile)
                    file = open(lfile,'w')
                    file.writelines([line1, line2, line3, line4, line5, line6, line7, line8, line9]) 
                    file.close()
```
