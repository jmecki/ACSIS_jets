```python
from netCDF4 import Dataset
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import glob
import os

import numpy as np
```

```python
figdir  = 'jet_figures/211210/'

def plot_Natl(plot_data,title,tcol,figfile,vmin,vmax,cmap,cbar):
    fig   = plt.figure(figsize=(15,10))
    fsize = 16
    ax    = plt.axes(projection=ccrs.PlateCarree())
    ax.set_extent([-95,75,0,80])
    ax.add_feature(cfeature.NaturalEarthFeature('physical', 'land', '50m', edgecolor='face', facecolor='0.5'))
    ax.coastlines()
    gl=ax.gridlines(draw_labels=True)
    gl.top_labels = False
    gl.left_labels = False
    ax.coastlines()
    gl.xlabel_style = {'fontsize': fsize}
    gl.ylabel_style = {'fontsize': fsize}

    plt.pcolormesh(lon,lat,plot_data,transform=ccrs.PlateCarree(),vmin=vmin,vmax=vmax,cmap=cmap)
    cb = plt.colorbar(extend=cbar,orientation='horizontal')
    cb.ax.tick_params(labelsize=fsize)
    cb.set_label('sea level (m)',size=fsize)
    plt.title(title,fontsize=fsize,color=tcol)
    
    #plt.show()

    fig.savefig(figfile,dpi=100,bbox_inches='tight', pad_inches=0.1)
    
    return fig
```

```python
var = 'zos'
datadir = '/gws/nopw/j04/acsis/jmecking/CMIP6/'
#files = glob.glob((datadir + '*/*/*/'+ var + '_*_*_????????_??_??????-??????_SC.nc'))
files = glob.glob((datadir + '*/historical/*/'+ var + '_*EN4*.nc'))

nf = len(files)

# Read in data similar to all models:
ncid = Dataset(files[0],'r')
lon = ncid.variables['lon'][:]
lat = ncid.variables['lat'][:]
ncid.close()
    
    
nx = len(lon)
ny = len(lat)
```

```python
climm = 2
vminm = -climm*1.05
vmaxm =  climm*1.05
cmapm = plt.get_cmap('seismic',21)

climd = 0.6
vmind = -climd*1.05
vmaxd =  climd*1.05
cmapd = plt.get_cmap('seismic',21)

vminr = 0
vmaxr = 0.4
cmapr = plt.get_cmap('YlOrRd',20)

climdr = 0.2
vmindr = -climdr*1.05
vmaxdr =  climdr*1.05
cmapdr = plt.get_cmap('seismic',21)

vminmon = -0.5
vmaxmon = 12.5
cmapmon = plt.get_cmap('twilight_shifted',13)
```
