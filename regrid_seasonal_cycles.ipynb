```python
import jCMIP as jc
import os
import shutil
import glob

Clist = jc.readList('../CMIP6list')
```

```python
# Generates namelist for regridding data (Model to EN4) (Ocean):

# Generate list of files to regrid:
var    = 'zos'

datadir = '/gws/nopw/j04/acsis/jmecking/CMIP6/'
files = glob.glob((datadir + '*/*/*/'+ var + '_*_*_????????_??_??????-??????_SC.nc'))


for infile in files:
    print(infile)
    Model = Clist[infile.split('/')[-1].split('_')[1]]
    EXP   = infile.split('/')[-1].split('_')[2]
    ENS   = infile.split('/')[-1].split('_')[3]
    gtype = infile.split('/')[-1].split('_')[4]
    dates = infile.split('/')[-1].split('_')[5]

    # Check if regridded file exists:
    regridfile = (datadir + Model.name + '/' + EXP + '/' + ENS + '/' + var + '_' + Model.name + '_' + EXP + '_' + ENS + '_' + gtype + '_' + dates + '-EN4_SC.nc')
    print(regridfile)
    if not (os.path.exists(regridfile)):
        print(Model.Omeshmask)
        if Model.Omeshmask != 'unknown':
            # create namelist for computation:
            namefile = (datadir + '/regrid_files/' + var + '_' + Model.name + '_' + EXP + '_' + ENS + '.namelist')
            shutil.copyfile(('/home/users/jmecking001/jpython/ACSIS_jets/template.namelist'),namefile)

            # Read in file:
            with open(namefile, 'r') as file:
                data = file.readlines()

                # Edit file (source information):
                data[74] = ('csource    = \'' + Model.name + '_' + EXP + '_' + ENS + '\'\n')
                data[75] = ('ivect      = 0\n')
                data[76] = ('l_reg_src  = .false.\n')
                data[77] = ('cf_src     = \'' + infile + '\'\n')
                data[78] = ('cv_src     = \'' + var + '\'\n')
                data[79] = ('cv_t_src   = \'time\'\n')
                data[80] = ('cf_x_src   = \'' + Model.Omeshmask + '\'\n')
                data[81] = ('cv_lon_src = \'tlon\'\n')
                data[82] = ('cv_lat_src = \'tlat\'\n')
                data[83] = ('cf_lsm_src = \'' + Model.Omeshmask + '\'\n')
                data[84] = ('cv_lsm_src = \'tmask\'\n')
                data[85] + ('ewper_src  = 0')

                data[88] = ('cf_z_src  = \'' + Model.Omeshmask + '\'\n')
                if Model.name == 'IPSL-CM6A-LR':
                    data[89] = ('cv_z_src  = \'olevel\'\n')
                else:
                    data[89] = ('cv_z_src  = \'lev\'\n')

                data[91] = ('ctype_z_src = \'z\'\n')

                # Edit file (target information):
                data[152] = ('ctarget    = \'EN4\'\n') 
                data[153] = ('l_reg_trg  = .true.\n')
                data[154] = ('cf_x_trg   = \'/home/users/jmecking001/DATADIR/NEMO/vol4/jmecking001/BAC/CMIP6/EN4/EN.4.2.1.f.analysis.g10.1970-1999_sigma0.nc''\'\n')
                data[155] = ('cv_lon_trg = \'lon\'\n')
                data[156] = ('cv_lat_trg = \'lat\'\n')
                data[157] = ('cf_lsm_trg = \'missing_value\'\n')
                data[158] = ('cv_lsm_trg = \'salinity\'\n')
                data[159] = ('ewper_trg  = 0\n')

                data[162] = ('cf_z_trg = \'/home/users/jmecking001/DATADIR/NEMO/vol4/jmecking001/BAC/CMIP6/EN4/EN.4.2.1.f.analysis.g10.1970-1999_sigma0.nc''\'\n')
                data[163] = ('cv_z_trg = \'depth\'\n')
                data[164] = ('ctype_z_trg = \'z\'\n')

                # Edit file (interpolation information):
                data[239] = ('cmethod     = \'bilin\'\n')

                data[241] = ('idrown      = 50,20\n')

                data[246] = ('jplev       = 1\n')

                # Edit file (output file information):
                data[291] = ('cv_out    = \'' + var + '\'\n')

                data[294] = ('cv_t_out  = \'time\'\n')

                data[296] = ('cextra    = \'' + dates + '\'\n')
                data[297] = ('lmout     = .true.\n')

                if Model.name == 'IPSL-CM6A-LR':
                    data[303] = ('cv_z_out  = \'olevel\'\n')
                else:
                    data[303] = ('cv_z_out  = \'lev\'\n')


                # Write out file:
                with open(namefile, 'w') as file:
                    file.writelines( data )                   
```
